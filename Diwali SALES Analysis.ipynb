```python
#importing python libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# IMPORT CSV FILE
df=pd.read_csv('/sales_data.csv',encoding="latin1")
df
```

```python
# INFORMATION ABOUT ALL DATASET
df.info()
```

```python
#finding row and columns
df.shape
```

```python
df.isnull().sum()
```

```python
#DROP UNRELATED/BLOCK COLUMNS
df.drop(columns=['Status','unnamed1'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Amount"].mean()
```

```python
# df.fillna(inplace=True) # fill
df.fillna(df["Amount"].mean(),inplace=True)
# df.head()
# df.isnull().sum()
```

```python
df.columns
```

```python
# #rename column
df.rename(columns={"Marital_Status":"Shadi"})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df[[ 'Cust_name', 'Product_ID', 'Gender']].describe()
```

#Exploratory Data Analysis
#Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df,color='red')

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# ARRANGING IN ACENDENDING ORDER BY GENDER INCLUDING THEIR SUM
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
# PLOTTING A BAR CHART FOR GENDER VS TOTAL AMOUNT
Sale_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount', data=Sale_gen)
plt.show()

```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)    #Ye function bar chart ke upar labels dikhata hai.
```

```python
# Total amount vs Age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#sales_age
sns.barplot(x='Age Group',y='Amount',data=sales_age,palette='Greens')
plt.show()
```

```python
#STATE
#total no.of orders from top 10 states
sales_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#sales_states
sns.set(rc={'figure.figsize':(15,5)})   # Ye Seaborn/Matplotlib plot ka size set karne ke liye use hota hai.
                                        # rc- ek predefined parameter hai jisse runtime configuration pass ki jati hai. isko change nhi kar skte

sns.barplot(data=sales_states, x='State', y='Orders', hue='Orders',color='blue')
plt.xticks(rotation=50)
plt.show()
```

```python
#total amount /sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})     #sns.set() Seaborn plots ke overall look aur style ko set karne ke liye use hota hai.
                                          # Matlab, ye default appearance change kar deta hai, jaise:
                                          # Figure ka size
                                          # Background style
                                          # Grid lines
                                          # Font size aur color palette

sns.barplot(data = sales_state, x = 'State',y= 'Amount',hue='Amount',palette='Greens')
plt.xticks(rotation=45)
plt.show()
```

```python
#MARITAL STATUS
ax=sns.countplot(data=df,x='Marital_Status',color="red",hue='Marital_Status')

sns.set(rc={'figure.figsize':(12,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# MARITAL STATUS VS AMOUNT BY GENDER

sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
#OCCUPATOPN
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
  plt.xticks(rotation=60)
  plt.show()
```

# OCCUPATION VS AMOUNT


```python
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_occ, x = 'Occupation',y= 'Amount',hue='Amount')

plt.title('OCCUPATION VS AMOUNT',fontsize=20)

plt.xticks(rotation=40)
plt.show()
```

# PRODUCT CATEGORY

```python
#PRODUCT CATEGORY
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=45)
plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector


# AMOUNT VS PRODUCT CATEGORY

```python
#AMOUNT VS PRODUCT CATEGORY
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount',palette="Greens")

plt.xticks(rotation=45)
plt.show()
```
